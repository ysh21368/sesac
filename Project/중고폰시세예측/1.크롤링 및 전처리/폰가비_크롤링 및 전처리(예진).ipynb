# 폰가비 크롤링

```python
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager # 알아서 버전을 맞춰서 다운받아옴.
```

```python
times=[]
price=[]
iphone=[]
brandnew=[]
tiny_d=[]
frame_d=[]
lcd_d=[]
screen_d=[]
function_d=[]
driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()))
url='https://fongabi.com/fon_2/deal_live/'
driver.get(url)
for i in range(1,3000):
    driver.find_element(By.CSS_SELECTOR,'#con_wrap > div.list > div:nth-child(' +str(i)+ ') > a ')
    name=driver.find_element(By.CSS_SELECTOR,'#con_wrap > div.list > div:nth-child(' +str(i)+ ') > a > h4')
    #driver.send_keys()
    print(i)
    if '아이폰' in name.text:
        iphone.append(name.text)
        ele=driver.find_element(By.CSS_SELECTOR,'#con_wrap > div.list > div:nth-child(' +str(i)+ ') > a > span')
        times.append(ele.text)
        ele=driver.find_element(By.CSS_SELECTOR,'#con_wrap > div.list > div:nth-child('+str(i)+') > a > p')
        price.append(ele.text)
        ele=driver.find_element(By.CSS_SELECTOR,'#con_wrap > div.list > div:nth-child(' +str(i)+ ') > a > h4')
        ele.click()
        time.sleep(1)
        data=driver.find_elements(By.CSS_SELECTOR,'#phone_condition > dd > table > tbody > tr > td:nth-child(2)')
        brandnew.append(data[0].text)
        tiny_d.append(data[1].text)
        frame_d.append(data[2].text)
        lcd_d.append(data[3].text)
        screen_d.append(data[4].text)
        function_d.append(data[5].text)
        #grading.append(data[6].text)
        driver.back()
        time.sleep(0.3*(i//10+1))
    if i%10==0:
        element = driver.find_element(By.CSS_SELECTOR,'body')
        element.send_keys(Keys.END)
        time.sleep(1)
        driver.find_element(By.CSS_SELECTOR,'#btn_review_more').click()
        time.sleep(2)
```

## [가격]과 [기종]에서 Feature 분리하기

```python
# 가격에서 등급과 가격으로 컬럼 분리하기.
abc=[]
pri=[]
for i in data['가격']:
    a = i.split(' ')
    abc.append(a[0].split('\n')[0])
    pri.append(a[1])
    
data['등급'] = abc
data['가격2'] = pri

#기종에서 아이폰, 기종 넘버, 용량 분리하기
num=[]
gb=[]
name=[]
for i in data['기종']:
    a = i.split(' ')
    num.append(a[1])
    gb.append(a[-1])
    name.append(a[0])

data['num'] = num
data['gb'] = gb
data['name'] = name
```

```python
# 기종에서 분류명 세분화해서 [기종2]로 Feature 분류하기
data.loc[data['기종'].str.contains('프로') & ~data['기종'].str.contains('맥스'), '기종2'] = '프로'
data.loc[data['기종'].str.contains('맥스'), '기종2'] = '프로맥스'
data.loc[data['기종'].str.contains('Max'), '기종2'] = 'Max'
data.loc[data['기종'].str.contains('플러스'), '기종2'] = '플러스'
data.loc[data['기종'].str.contains('미니'), '기종2'] = '미니'
data['기종2'] = data['기종2'].fillna("없음")
```

```python
# data를 카피해서 '기능불량'의 컨탠츠를 각각의 컬럼으로 0,1로 전처리하기 위해 복사해서 새로운 데이터프레임 만들기
data2 = data.copy()
data2
```

```python
# ['기능불량']에 있는 중구난방 데이터들을 띄어쓰기를 없애고(공백제거) 콤마로 분류하기
data3 = pd.DataFrame(data['기능불량'].str.replace(' ','').str.split(','))
data3
```

```python
# 위의 data3속 분류된 데이터들을 각각의 컬럼으로 한 번에 전처리하기.
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer(sparse_output=True)

df = data3.join(
            pd.DataFrame.sparse.from_spmatrix(
                mlb.fit_transform(data3.pop('기능불량')), #pop되어서 차후 기능불량 시리즈만 있는 data3의 기능불량 컬럼은 사라짐.
                index=data3.index,
                columns=mlb.classes_))
```
