```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
cetizen=pd.read_csv('phone_complete.csv',encoding='euc-kr')
```

## 피처별 가격 분포 - 세티즌

```python
cetizen.columns
```

```python
features=['제품상태', '구성품', '기변상태', '선택약정', '보증기간', '용량', '분류1', '분류2']
```

```python
for f in features:
    plt.figure(figsize=(5,5))
    plt.grid()
    sns.boxplot(x=f,y='가격',data=cetizen.sort_values(f))
    plt.xticks(rotation=45)

    current_values = plt.gca().get_yticks()
    plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in current_values])

    plt.title('아이폰가격 분포')
    plt.show()
```

## 피처별 가격 분포 - 폰가비

```python
gabi=pd.read_csv('gabi_final.csv',encoding='euc-kr')
```

```python
gabi['시리즈'] = '아이폰' +' '+ gabi['시리즈']
```

```python
plt.figure(figsize=(17,10))
plt.grid()
sns.countplot(x='시리즈',data=gabi,
                order=gabi['시리즈'].value_counts().index)
plt.xticks(rotation=45)

current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in current_values])

plt.title('시리즈별 매물 건수 (폰가비)')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.grid()
sns.boxplot(x='개봉여부',y='가격',data=gabi,
                order=gabi['개봉여부'].value_counts().index)
plt.xticks(rotation=45)

current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in current_values])

plt.title('개봉여부별 가격분포 (폰가비)')
plt.show()
```

```python
features=['스피커','진동', '마이크', '배터리성능저하', '생체인식','카메라', '터치', '계정잠김']
```

```python
temp=gabi.groupby('마이크')['가격'].median().reset_index()
sns.barplot(temp['마이크'],temp['가격'])
```

```python
fig, ax = plt.subplots(figsize=(18,6),ncols=4)
i=0
for f in  features[:4]:
    temp= gabi.groupby(f)['가격'].median().reset_index()
    sns.barplot(temp[f],temp['가격'],ax=ax[i])
    
    
    i=i+1
```

```python
fig, ax = plt.subplots(figsize=(18,6),ncols=4)
i=0
for f in  features[4:]:
    temp= gabi.groupby(f)['가격'].median().reset_index()
    g=sns.barplot(temp[f],temp['가격'],ax=ax[i])
    
    axs = g
    #annotate axis = seaborn axis 
    for p in axs.patches: 
        axs.annotate("%.0f" % p.get_height(), (p.get_x() + p.get_width()/2., p.get_height() - 30), 
           ha='center', va='center', fontsize=10, color='black', xytext=(0, 10), 
           textcoords='offset points') 
    i=i+1
```

```python
fig, ax = plt.subplots(figsize=(18,6),ncols=4)
sns.boxplot(x='배터리성능저하',y='가격',data=gabi,
                order=gabi['배터리성능저하'].value_counts().index ,ax=ax[0])
sns.boxplot(x='생체인식',y='가격',data=gabi,
                order=gabi['생체인식'].value_counts().index ,ax=ax[1])
sns.boxplot(x='스피커',y='가격',data=gabi,
                order=gabi['스피커'].value_counts().index ,ax=ax[2])
sns.boxplot(x='계정잠김',y='가격',data=gabi,
                order=gabi['계정잠김'].value_counts().index ,ax=ax[3])
```
